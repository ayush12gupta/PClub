**Defining X and Y **

```
X=np.random.rand(50,1)
#print(X)
y=np.random.rand(50,1)
#m,n=X.shape
X = np.hstack((np.ones((50,1)),X,X**2,X**3,X**3,X**4))
m,n=X.shape
w = np.random.rand(n,1)
```

```
plt.scatter(X[:,1],y)
plt.show()
```

**Normalising X**

```
#Normalising 
for i in range(1,n):
    X[:,i] = (X[:,i] - np.mean(X[:,i], axis=0)) / np.std(X[:,i], axis=0)
```

**Loss Functon for finding Error**

```

# loass function
def loss(y_pred,y):
    J = np.mean(np.square(y-y_pred))/2
    return J    
```

**Function fot training and doing Gradient Descent**

```
def grad_desc(X,y,w,epochs=500,lr=0.01):
     error_hist=np.zeros(epochs)
    # w = get_linear_model(X,y)
     for i in range(epochs):
        w = w - lr*(X.T@((X @ w) - y))/m
        err = loss((X@w),y)
        error_hist[i] = err
        print("Error: ",err," Epoch: ",i)

     return w , error_hist   
```

```
w,error = grad_desc(X,y,w)
```
